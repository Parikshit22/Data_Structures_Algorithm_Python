```python
class PriorityQueue:
    def __init__(self,):
        self.arr = []
        
    def getmax(self):
        return self.arr[0]
    
    def getsize(self):
        return len(self.arr)
    def is_empty(self):
        return self.getsize==0
    
    def insert(self,value):
        self.arr.append(value)
        idx = len(self.arr)-1
        
        while idx>0:
            pr_idx = (idx-1)//2
            if self.arr[pr_idx]<self.arr[idx]:
                self.arr[pr_idx],self.arr[idx] = self.arr[idx],self.arr[pr_idx]
            else:
                break
            idx = pr_idx
    def remove_min(self):
        ab = self.arr[0]
        self.arr[0] = self.arr[-1]
        self.arr.pop()
        idx = 0
        
        while True:
            ch1 = 2*(idx)+1
            ch2 = 2*(idx)+2
            if ch2<len(self.arr):
                if self.arr[ch1]>self.arr[ch2]:
                    maxm = ch1
                else:
                    maxm = ch2
            else:
                maxm = ch1
            if self.arr[maxm]>self.arr[idx]:
                self.arr[maxm],self.arr[idx] =self.arr[idx], self.arr[maxm]
            else:
                break
            idx = maxm
            val = ((len(self.arr)/2)-1)
            if idx > val:
                break
        
        return ab
    
            
                  
myPq = PriorityQueue()
curr_input = [int(ele) for ele in input().split()]
choice = curr_input[0]
i=1
while choice != -1:
    if choice == 1:
        element = curr_input[i]
        i+=1
        myPq.insert(element)
    elif choice == 2:
        print(myPq.getmin())
    elif choice == 3:
        print(myPq.remove_min())
    elif choice == 4:
        print(myPq.getsize())
    elif choice == 5:
        if myPq.is_empty():
            print('true')
        else:
            print('false')
        break
    else:
        pass
    choice = curr_input[i]
    i+=1
        
    
```
