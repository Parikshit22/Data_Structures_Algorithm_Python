```python

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class QueueUsingLL:
  
### Implement These Functions Here    
    def __init__(self):
        self.__head = None
        self.__tail = None
        self.__curr = 0
        
    def enqueue(self,data):
        new = Node(data)
        self.__curr+=1
        if self.__head is None:
            self.__head = new
            self.__tail = new
            return
        self.__tail.next = new
        self.__tail  = self.__tail.next
        return
        
        
    def dequeue(self):
        if self.__head is None:
            return -1
        self.__curr-=1
        val = self.__head.data
        if self.__head==self.__tail:
            self.__head=None
            self.__tail = None
            return val
        self.__head = self.__head.next
        return val
        
        
    
    def front(self):
        if self.__head is None:
            return -1
        return self.__head.data
        
    def isEmpty(self):
        if self.__head is None:
            return True
        return False
    
    def getSize(self):
        return self.__curr
    
q = QueueUsingLL()
li = [int(ele) for ele in input().split()]
i=0
while i<len(li):
    choice = li[i]
    if choice == -1:
        break
    elif choice == 1:
        q.enqueue(li[i+1])
        i+=1
    elif choice == 2:
        ans = q.dequeue()
        if ans!=0:
            print(ans)
        else:
            print(-1)
    elif choice == 3:
        ans = q.front()
        if ans!=0:
            print(ans)
        else:
            print(-1)
    elif choice == 4:
        print(q.getSize())
    elif choice == 5:
        if(q.isEmpty()):
            print('true')
        else:
            print('false')
    i+=1






```
